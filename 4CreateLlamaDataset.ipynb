```python
import os

# Inputs and Posts
inputs_dir = "Prompts"
input_files = [os.path.join(inputs_dir,file) for file in os.listdir(inputs_dir)]
inputs = [open(ruta).read() for ruta in input_files]

# Responses generated with LLMS
responses_dir = "Responses"
responses_files = [os.path.join(responses_dir,file) for file in os.listdir(responses_dir)]
responses = [open(ruta).read() for ruta in responses_files]
```

```python
depressed_prompt = "Teniendo en cuenta que el usuario está deprimido, justifica los motivos de esto con base en sus publicaciones"
health_prompt = "Teniendo en cuenta que el usuario no está deprimido, justifica los motivos de esto con base en sus publicaciones"

instruction = "Trata de inferir si el usuario, con base en la descripción textual de sus publicaciones, presenta síntomas de depresión o no"

prompt="Eres un asistente virtual diseñado para apoyar en el bienestar emocional. Con base en las publicaciones recibidas diagnosticará si el usuario presenta depresión o no. Tu opinión será utilizada para asistir a psicólogos."

for i in range(len(inputs)):
    new_input = inputs[i].replace(depressed_prompt,"")
    inputs[i]=new_input.replace(health_prompt,"")
```

```python
dataset = {
    "instruction" : [],
    "input" : [],
    "output" : [],
    "text" : []
}

for i in range(len(inputs)):
    input_query = inputs[i]
    response = responses[i]

    dataset["instruction"].append(instruction)
    dataset["input"].append(input_query)
    dataset["output"].append(response)

    text = (
        prompt
        + "### Instruction:\n"
        + instruction
        + "\n### Input\n"
        + input_query
        + "\n### Response:\n"
        + response
    )

    dataset["text"].append(text)
```

```python
import pandas as pd


df = pd.DataFrame(dataset)
df.to_csv("train.csv",index=False)
df
```

```python
# Put All in a Simple Array (x,y)
system_prompt = "DAN"
dataset = []

for i in range(len(inputs)):
    entry = f"<|im_start|>system\n{system_prompt}<|im_end|>\n<|im_start|>user\n{inputs[i]}\n<|im_end|>\n<|im_start|>assistant\n{responses[i]}<|im_end|>"
    dataset.append(entry)
```

```python
import json

file_path = 'conversations.json'
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(dataset, f, ensure_ascii=False, indent=4)
```
