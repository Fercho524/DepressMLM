```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

palabras_frecuencias_desconocida = {
    "Buena persona": 2,
    "Paciente": 1,
    "Tranquilo": 3,
    "Divertido": 3,
    "Noble": 2,
    "Gran futuro": 1,
    "Inocente": 1
}

wordcloud_desconocida = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    colormap='magma',
    relative_scaling=0.5
).generate_from_frequencies(palabras_frecuencias_desconocida)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_desconocida, interpolation='bilinear')
plt.axis('off')
plt.title("Nube de Palabras: Área Desconocida")
plt.show()
```
